```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt


new_data['DATE OCC'] = pd.to_datetime(new_data['DATE OCC'])
new_data = new_data.sort_values('DATE OCC')


daily_crimes = new_data.groupby('DATE OCC').size().reset_index(name='crime_count')
daily_crimes = daily_crimes.set_index('DATE OCC')

model = ARIMA(daily_crimes['crime_count'], order=(1,1,1))
results = model.fit()


forecast = results.forecast(steps=30)


plt.figure(figsize=(12, 6))
plt.plot(daily_crimes.index, daily_crimes['crime_count'], label='Observed')
plt.plot(forecast.index, forecast, color='red', label='Forecast')
plt.title('Crime Trend Forecast')
plt.xlabel('Date')
plt.ylabel('Number of Crimes')
plt.legend()
plt.show()
```

#2.

Null Hypothesis: The incidence of violent crimes does not statistically differ on weekends and weekdays

```python
import seaborn as sns
from scipy import stats
data['DATE OCC'] = pd.to_datetime(data['DATE OCC'])
data['Day of Week'] = data['DATE OCC'].dt.day_name()
list_of_crimes = pd.read_csv('Book1.csv')
crimes = list_of_crimes['Crm Cd Desc'].tolist()
data_violent = data[data['Crm Cd Desc'].isin(crimes)]
data_violent['Day Type'] = data_violent['Day of Week'].apply(lambda x: 'Weekend' if x in ['Saturday', 'Sunday'] else 'Weekday')
crime_counts = data_violent['Day Type'].value_counts()
chi2_stat, p_val = stats.chisquare(crime_counts)
print(f'Hypothesis 1: Chi-Squared Statistic = {chi2_stat}, p-value = {p_val}')
plt.figure(figsize=(9, 7))
sns.countplot(x='Day Type', data=data_violent, palette='muted',hue='Day Type')
plt.title('Crimes on Weekends vs Weekdays')
plt.xlabel('Day Type')
plt.ylabel('Count of Crimes')
plt.show()
```

The Chi Square test results suggests that the crime incidence on weekends are more likely. 
Here, we are rejecting the null Hypothesis that crime is equally distributed on days of week

Null Hypothesis: There is statistical no signifcant difference in the frquency of crimes on the age of victims

```python

crime_age = pd.crosstab(data['Vict Age'], data['Crm Cd Desc'])


chi2, p, dof, expected = chi2_contingency(crime_age)


print(f'Hypothesis : Chi-Squared Statistic = {chi2}, p-value = {p}')
```

We reject the null hypothesis based on the test result. The conclusion is that victims age has significant impact on crime occurence

```python
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Find outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers


outliers = detect_outliers_iqr(data, 'Vict Age')


print(outliers)
```

```python
#Converting 'DATE OCC' to datetime format
data['DATE OCC'] = pd.to_datetime(new_data['DATE OCC'], format='%d-%m-%Y %H:%M', errors='coerce')

#Creating a 'YearMonth' column in the crime dataset and converting it to timestamp
data['YearMonth'] = data['DATE OCC'].dt.to_period('M')  # Convert to YYYY-MM Period
data['YearMonth'] = data['YearMonth'].apply(lambda x: x.to_timestamp())  # Convert Period to Timestamp

#Creating a 'YearMonth' column in the CPI dataset without explicit format
cpi_df['YearMonth'] = pd.to_datetime(cpi_df['YearMonth'], errors='coerce') 

data['YearMonth'] = data['YearMonth'].dt.to_period('M').dt.to_timestamp()
cpi_df['YearMonth'] = cpi_df['YearMonth'].dt.to_period('M').dt.to_timestamp()

#Filtering relevant columns from the dataset
demographic_cols = ['YearMonth', 'Vict Age', 'Vict Sex', 'Vict Descent', 'Crm Cd Desc']
demographic_data = data[demographic_cols]

#Group by demographics and crime descriptions, including YearMonth
demographic_group = demographic_data.groupby(['YearMonth', 'Vict Age', 'Vict Sex', 'Vict Descent', 'Crm Cd Desc']).size().reset_index(name='Crime Count')



#Drop NaN values if any
demographic_group = demographic_group.dropna(subset=['YearMonth'])
cpi_df = cpi_df.dropna(subset=['YearMonth'])

#Merging with CPI data
merged_df = pd.merge(demographic_group, cpi_df, on='YearMonth', how='inner')

#Visualizing crime distribution by demographic factors
plt.figure(figsize=(10, 6))
sns.histplot(data=demographic_group, x='Vict Age', hue='Vict Sex', multiple='stack', bins=30)
plt.title('Crime Distribution by Victim Age and Gender')
plt.xlabel('Victim Age')
plt.ylabel('Crime Count')
plt.show()
```

```python
#Analyzing and visualizing specific crime types by gender

#Fetching the top 3 crime types based on the total count
top_crime_types = demographic_group['Crm Cd Desc'].value_counts().nlargest(3).index

#Filtering the demographic_group to include only the top 3 crimes
top_crime_data = demographic_group[demographic_group['Crm Cd Desc'].isin(top_crime_types)]

#Visualize crime types by gender
plt.figure(figsize=(12, 6))
sns.countplot(data=top_crime_data, x='Crm Cd Desc', hue='Vict Sex', order=top_crime_types)
plt.xticks(rotation=45)
plt.title('Top 3 Crime Types by Victim Gender')
plt.xlabel('Crime Description')
plt.ylabel('Crime Count')
plt.show()
```
